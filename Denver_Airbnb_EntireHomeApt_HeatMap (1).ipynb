```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
# Importing listings CSV and storing it in a dataframe: den_data
all_den_data = pd.read_csv(r"C:\Users\bdmul\Documents\Coding applications\Python Projects\STRs\listings.csv")

# Viewing den_data dataframe
all_den_data 
```

```python
# Dropping unnessecary columns   
denver_data = all_den_data.drop(['neighbourhood_group', 'last_review', 'reviews_per_month', 'number_of_reviews', 'calculated_host_listings_count',
                   'availability_365', 'minimum_nights', 'host_name', 'host_id', 'name', 'id'], axis=1)

# setting index to 'neighbourhood'
#denver_data = denver_data.set_index('neighbourhood')
```

```python
# Splitting Entire home/apt into own dataframe
grouped = denver_data.groupby(denver_data.room_type)
entire_home_apt = grouped.get_group("Entire home/apt")
entire_home_apt 
```

```python
hmap = folium.Map(location=[39.742043, -104.991531], zoom_start=11)

hm_wide = HeatMap(list(zip(entire_home_apt.latitude.values, entire_home_apt.longitude.values)),
                   min_opacity=0.2,
                   radius=17, blur=15, 
                   max_zoom=1)

geojson = r'C:/Users/bdmul/PycharmProjects/folium_projects/neighbourhoods.geojson'

g = folium.GeoJson(geojson, name="geojson").add_to(hmap)

# GeoJsonTooltip(fields=[‘CNTY_NM’, ‘census-pop-2015’, ‘census-md-income-2015’],
folium.GeoJsonTooltip(fields=["neighbourhood"]).add_to(g)



hmap.add_child(hm_wide)

hmap
```
